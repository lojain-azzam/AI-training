# Dot Product 
The dot product, also known as the `scalar product` or `inner product`, is a fundamental operation in linear algebra and vector mathematics. It is a binary operation that takes two vectors as input and produces a scalar (single number) as output

```python
from IPython.display import Image
Image(filename='images/dotproduct.jpg', width=750)
```

```python
# Dot Product Operation using NumPy

import numpy as np


vector1 = np.array([1, 2, 3])
vector2 = np.array([4, 5, 6])

# Calculate dot product using NumPy
dot_product_numpy = np.dot(vector1, vector2)



print(f"Dot product using NumPy: {dot_product_numpy}")
```

### Question:
What is the dot product operation, and how is it implemented in NumPy?
### Answer: 
The dot product(scalar product or inner product) operation in linear algebra and vector mathematics. It is used to calculate the product of two vectors, resulting in a scalar value. the dot product operation is implemented using the np.dot() function.

The **dot product operation** multiplies the corresponding elements of the two vectors and then sums up the results. In this case, the dot product of the vectors [1, 2, 3] and [4, 5, 6] is `1*4 + 2*5 + 3*6 = 4 + 10 + 18 = 32.`



## Cosine Similarity
Cosine similarity is a measure of similarity between two non-zero vectors, often used in information retrieval, text mining, and machine learning. It measures the cosine of the angle between two vectors, which is a value between `-1 and 1.`

```python
from IPython.display import Image
Image(filename='images/cosine.png', width=750)
```

```python

# Cosine Similarity Implementation

# Implement cosine similarity using the dot product
def cosine_similarity(v1, v2):
    dot_product = np.dot(v1, v2)
    norm_v1 = np.linalg.norm(v1)
    norm_v2 = np.linalg.norm(v2)
    
    cosine_sim = dot_product / (norm_v1 * norm_v2)
    
    return cosine_sim


# Example vectors
vector_a = np.array([1, 2, 3])
vector_b = np.array([4, 5, 6])

similarity = cosine_similarity(vector_a, vector_b)
print(f"Cosine similarity: {similarity}")
```

**NOTE :**

`it calculates the magnitudes (lengths) of the two vectors using np.linalg.norm(v1) and np.linalg.norm(v2).`

`norm_v1 = sqrt(1^2 + 2^2 + 3^2) = sqrt(1 + 4 + 9) = sqrt(14)  3.742` . 

`norm_v2 = sqrt(4^2 + 5^2 + 6^2) = sqrt(16 + 25 + 36) = sqrt(77)  8.775`.

```python
# Word Embedding Example with Advanced Model

!pip install sentence-transformers

import numpy as np
from sentence_transformers import SentenceTransformer

# Load a more advanced pre-trained model
model = SentenceTransformer('all-MiniLM-L6-v2')

def get_sentence_embedding(sentence):
    return model.encode(sentence)

def cosine_similarity(v1, v2):
    return np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))

def dot_product_similarity(v1, v2):
    return np.dot(v1, v2)

# Function to compare sentences
def compare_sentences(sentence1, sentence2):
    embedding1 = get_sentence_embedding(sentence1)
    embedding2 = get_sentence_embedding(sentence2)

    cosine_sim = cosine_similarity(embedding1, embedding2)
    dot_sim = dot_product_similarity(embedding1, embedding2)

    print(f"Sentence 1: '{sentence1}'")
    print(f"Sentence 2: '{sentence2}'")
    print(f"Cosine similarity: {cosine_sim:.4f}")
    print(f"Dot product similarity: {dot_sim:.4f}")
    print()

# Example usage
compare_sentences("I love programming", "I enjoy coding")
compare_sentences("dogs are better than cats", "cat is funny!")
compare_sentences("I love pizza", "I hate vegetables")
compare_sentences("im a doctor", "sell some phones!")


# Task for users:
print("Now it's your turn! Try comparing different sentences and have fun exploring similarities!")
print("1. Compare sentences with similar meanings but different words.")

compare_sentences("The sun is shining brightly today", "It's a sunny day outside")
compare_sentences("My stomach is rumbling", "I'm hungry")
compare_sentences("I need to get some rest", "I'm tired")


print("2. Try sentences with the same words in different orders.")

compare_sentences("The dog chased the cat", "Chased the dog the cat.")
compare_sentences("She opened the door slowly", "Slowly she opened the door")
compare_sentences("The red car drove fast", "The car red drove fast.")


print("3. Experiment with sentences from different domains or languages.")

compare_sentences("TLa plume de ma tante est sur la table", "The pen of my aunt is on the table")
compare_sentences("", "My aunt's pen is on the table")
compare_sentences("My aunt's pen is on the table", "The Wi-Fi router transmits the internet signal")



print("4. Sentence for text :")

compare_sentences("lojain azzam ahmad banishamsa", "lojain azzam ahmad banishamsa")


# Reflection questions:
print("\nReflection questions:")
print("1. How do cosine similarity and dot product similarity yield identical results in the given examples? What does this reveal about their mathematical properties?")
print("2. What types of sentences seem to have high similarity? Low similarity?")



```

**Answer :**
When the vectors are normalized, the cosine similarity formula simplifies to just the dot product of the vectors:
`cosine_similarity(u, v) = dot(u, v) / (norm(u) * norm(v)) = dot(u, v) / (1 * 1) = dot(u, v)`
This mathematical property reveals that for **normalized vectors, cosine similarity and dot product similarity are equivalent measures of similarity**. They will provide the same relative ordering of similarity between vectors.

## Pseudo-inverse


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate correlated random points
mean = [0, 0]
cov = [[1, 0.8], [0.8, 1]]  # Covariance matrix with correlation
points = np.random.multivariate_normal(mean, cov, 100)
X = points[:, 0].T
y = points[:, 1].T

# Plot the data points
plt.scatter(X, y)
plt.xlabel('X')
plt.ylabel('y')
plt.title('Correlated Random Points')
plt.show()

# Create the design matrix [X 1] 
X_design = np.column_stack((X, np.ones_like(X)))

# Calculate the pseudo-inverse
X_pseudo_inv = np.linalg.pinv(X_design)

# Calculate the coefficients
coefficients = X_pseudo_inv.dot(y) #size of 2x1

# Print the coefficients
print(f"Estimated coefficients: {coefficients}")

# Plot the results
plt.scatter(X, y, label='Data points')
plt.plot(X, np.dot(X_design, coefficients), color='red', label='Linear model')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.title('Linear Regression using Pseudo-Inverse')
plt.show()
```

**NOTE**
* ` the pseudo-inverse is used to calculate the coefficients of the linear model that best fit the data.`
* `X_design = np.column_stack((X, np.ones_like(X))) matrix has a size of 100x2 where the first column contains the X values, and the second column is a column of ones.`
* `The coefficients of the linear regression model are calculated by performing a matrix multiplication between the pseudo-inverse X_pseudo_inv and the y variable.`
